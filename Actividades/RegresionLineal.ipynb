```python
# Librerías a utilizar
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
# Generar conjunto de datos
np.random.seed(0)
horas_estudio = np.random.rand(100, 1) * 10
calificaciones = 2 * horas_estudio + np.random.randn(100,1) * 2
```

```python
# Preprocesamiento de datos
train_data, test_data, train_labels, test_labels = train_test_split(horas_estudio, calificaciones, test_size=0.2, random_state=42)
```

```python
# Seleccionar el modelo a entrenar: Regresión lineal
modelo = LinearRegression()
```

```python
# Entrenamiento de modelo de Regresión Lineal
modelo.fit(train_data, train_labels)
```

```python
# Predicciones con el modelo entrenado
predicciones = modelo.predict(test_data)
```

```python
# Calcular el error cuadrático medio
error = mean_squared_error(test_labels, predicciones)
print(f'El error cuadrático medio es: {error}')
```

```python
# Gráfica de los datos originales
plt.scatter(train_data, train_labels, color='blue')
plt.title("Horas de Estudio vs Calificaciones")
plt.xlabel("Horas de Estudio")
plt.ylabel("Calificaciones")
plt.show()
```

```python
# Gráfica final con el modelo entrenado
plt.scatter(test_data, test_labels, color='red')
plt.plot(test_data, predicciones, color='blue')
plt.xlabel("Horas de Estudio")
plt.ylabel("Calificaciones")
plt.title("Modelo de Regresión Lineal")
plt.show()
```
